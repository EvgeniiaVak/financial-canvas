```python
ticker = yf.Ticker("AAPL")
df = ticker.history(start='2020-01-02', end='2021-01-01')
df.columns = map(str.lower, df.columns)
df
```

```python
mf = MultiCircle(df, columns=['open', 'close'], colors=['red', 'blue'])
mf.p.diamond('index', 'high', legend_label='close', size=9, source=mf.sources['main'][0])
show(mf.p)
```
