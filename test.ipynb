```python
from financial_canvas.figures.Candlestick import Candlestick
import yfinance as yf
from bokeh.io import output_notebook, show

output_notebook()
```

```python
ticker = yf.Ticker("AAPL")
df = ticker.history(start='2020-01-02', end='2021-01-01')
df.columns = map(str.lower, df.columns)
df
```

```python
candlestick = Candlestick(df, selected_from=df.index[0])
show(candlestick.p)
```
