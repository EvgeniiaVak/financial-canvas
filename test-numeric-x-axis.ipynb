```python
%reload_ext autoreload
%autoreload 2

from financial_canvas.figures.Line import Line
from financial_canvas import indicators_glyphs
import pandas as pd
import numpy as np
import yfinance as yf
from bokeh.layouts import column
from bokeh.io import output_notebook, output_file, show

output_notebook()
```

```python
ticker = yf.Ticker("AAPL")
df = ticker.history(start='2017-01-01', end='2020-12-31')
df.columns = map(str.lower, df.columns)
df
```

```python
lengths = []
avg_prices = []
for i in range(1, 1000):
    lengths.append(i)
    avg_prices.append(df.tail(i).close.mean())

mas = pd.DataFrame(data=avg_prices, index=lengths)
mas.columns = ['avg_price']
mas.index.name = 'length'
mas.index = -mas.index
mas.sort_index(inplace=True)
mas['open'] = None 
mas['open'] = mas['open'].astype('float64')
mas.at[-5, 'open'] = 150.0
mas.info()
```

```python
numeric_line = Line(mas, column_name='avg_price')
numeric_line.p.diamond('index', 'open', legend_label='open', size=9, source=numeric_line.sources['main'][0])
numeric_line.add_hover({'open': 'Open'})
# adding glyphs not supported
# numeric_line.p.y_range_resize_columns.append('open')


datetime_line = Line(df, column_name='close')

# output_file('some.html')
show(column(children=[numeric_line.p, datetime_line.p], sizing_mode='scale_both'))
```
